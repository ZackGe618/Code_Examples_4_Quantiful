12. I want to transform the above process into a function, and output them to a Data Frame
get_scripts() is designed for gluing the base link with each links, and extracting scripts texts from the  new url page. 

```julia
using DataFrames
```

```julia
function get_scripts(Script_link)
  response = HTTP.request("GET","$baseURL$Script_link") # we read the right page glueing together the base url and the link we got before
  parsed_Script_page = response.body |> String |> parsehtml
  sel_for_script = Selector(".full-script")
  text_script = eachmatch(sel_for_script,parsed_Script_page.root)
  script =  nodeText.(text_script)
  df = DataFrame(Link = Script_link, Scripts = script)
  return df
end
```
